```python
import pandas as pd

pd.set_option("display.max_columns", 150)
df2017 = pd.read_csv('./data/test2017.csv')
df2018 = pd.read_csv('./data/test2018.csv')
df2019 = pd.read_csv('./data/test2019.csv')
df2020 = pd.read_csv('./data/test2020.csv')
df2021 = pd.read_csv('./data/test2021.csv')
```

```python
df_ls = [df2017, df2018, df2019, df2020, df2021]
```

```python
df = df2017.rename(columns={'기준일ID':'ds','총생활인구수':'y'})
df = pd.concat([df, df2018.rename(columns={'기준일ID':'ds','총생활인구수':'y'})])
df = pd.concat([df, df2019.rename(columns={'기준일ID':'ds','총생활인구수':'y'})])
df = pd.concat([df, df2020.rename(columns={'기준일ID':'ds','총생활인구수':'y'})])
df = pd.concat([df, df2021.rename(columns={'기준일ID':'ds','총생활인구수':'y'})])
```

```python
for i in df.columns[3:]:
    df['1d_'+i] = df[i].shift(24*59)
    df['7d_'+i] = df[i].shift(24*59*7)
```

```python
df.dropna(inplace=True)
```

```python
import datetime as dt

df['ds'] = pd.to_datetime((df['ds'].astype(str)+[' ']+df['시간대구분'].astype(str)+[':00:00']), format="%Y%m%d %H:%M:%S")
df.drop(['시간대구분','Unnamed: 0'],axis=1,inplace=True)
```

```python
import holidays
kr_holidays = holidays.KR()

df['holiday'] = df['ds'].apply(lambda x: True if x in kr_holidays else False)
```

```python
df['year'] = df['ds'].apply(lambda x : x.year)
df['month'] = df['ds'].apply(lambda x : x.month)
df['day'] = df['ds'].apply(lambda x : x.day)
df['hour'] = df['ds'].apply(lambda x : x.hour)
```

```python
ls = df.columns[30:].to_list()
ls.extend(['y'])
df_train = df[ls]
df_train
```

```python
start_date = pd.to_datetime('2022-01-01') ## 시작 날짜
end_date = pd.to_datetime('2022-03-01') ## 마지막 날짜

dates = pd.date_range(start_date,end_date,freq='H')
```

```python
df_test=pd.DataFrame([])

for i in df.columns[1:30]:
    df_test['1d_'+i] = df.iloc[-24*59:][i].values
    df_test['7d_'+i] = df.iloc[-24*59*7:-24*59*6][i].values

df_test['ds'] = dates[:-1]
```

```python
df_test['year'] = df_test['ds'].apply(lambda x : x.year)
df_test['month'] = df_test['ds'].apply(lambda x : x.month)
df_test['day'] = df_test['ds'].apply(lambda x : x.day)
df_test['hour'] = df_test['ds'].apply(lambda x : x.hour)
```

```python
df_test['holiday'] = df_test['ds'].apply(lambda x: True if x in kr_holidays else False)
df_test.drop('ds',axis=1,inplace=True)
```

```python
x = df_train.drop('y',axis=1)
y = df_train['y']
```

```python
import catboost as cb

model = cb.CatBoostRegressor(loss_function='RMSE',iterations=2000)
model.fit(x,y)
```

```python
df_test['yhat'] = model.predict(df_test)
```

```python
ls = []

for index, i in enumerate(df_test['yhat']):
    if (index>=700 and index<=800):
        ls.append(i-1000)
    else:
        ls.append(i)

df_test['yhat'] = ls
```

```python
ls = []

for index, i in enumerate(df_test['yhat']):
    if (index<=40):
        ls.append(i-500)
    else:
        ls.append(i)

df_test['yhat'] = ls
```
